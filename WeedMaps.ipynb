```python
import pandas as pd
import re
```

```python
JuneDF = pd.read_csv('/Users/anishdulla/Desktop/WM_Feb15.csv')
```

```python
JuneDF
```

```python
JuneDF2 = JuneDF.loc[JuneDF['Status'] == 'Complete'].copy()
```

```python
JuneDF2
```

```python
del JuneDF2['Date/Time']
del JuneDF2['Type']
del JuneDF2['Source']
del JuneDF2['Zone/Location Name']
del JuneDF2['WMID']
del JuneDF2['Customer Name']
del JuneDF2['Item Quantity']
del JuneDF2['Order Items Total']
del JuneDF2['Excise Tax']
del JuneDF2['Sales Tax']
del JuneDF2['Service Fee']
del JuneDF2['Delivery Fee']
del JuneDF2['Discount Total']
del JuneDF2['Payment Type']
del JuneDF2['Discount Code']
del JuneDF2['Customer Notes']
```

```python
def regex(string):
    string2 = string[1:]
    return string2
```

```python
JuneDF2['Gross Sales'] = JuneDF2['Gross Sales'].apply(lambda x: regex(x))
```

```python
JuneDF2
```

```python
def RemoveComma(string3):
    string3 = string3.replace(',', '')
    return string3
```

```python
JuneDF2['Gross Sales'] = JuneDF2['Gross Sales'].apply(lambda x: RemoveComma(x))
```

```python
JuneDF2['Gross Sales'] = JuneDF2['Gross Sales'].astype(float)
```

```python
JuneDF3 = JuneDF2.groupby(['Listing Name']).agg({'Gross Sales': ['count', 'sum', 'mean']})
```

```python
JuneDF3 = JuneDF3.reset_index()
```

```python
JuneDF3.columns = JuneDF3.columns.to_flat_index()
```

```python
JuneDF3 = JuneDF3.rename(columns = {('Listing Name', ''): 'Listing', ('Gross Sales', 'sum'): 'Total Gross Sales', ('Gross Sales', 'mean'): 'Average Order Value', ('Gross Sales', 'count'): 'Total Orders'})
```

```python
def Round(floatie):
    return round(floatie, 2)
```

```python
JuneDF3['Average Order Value'] = JuneDF3['Average Order Value'].apply(lambda x: Round(x))
```

```python
JuneDF3['Total Gross Sales'] = JuneDF3['Total Gross Sales'].apply(lambda x: Round(x))
```
